```python
df = pd.read_csv("D:/aliem/ISRUC_Dataset/Demography_Details.csv")  
df
```

```python
df['Serial'] = df['File_Name'].apply(lambda x:x[-4:]) 
df 
```

```python
subs = ['subgroupI', 'subgroupII', 'subgroupIII']
all_paths = [] 
for sub in subs:
    direc = f"D:/aliem/ISRUC_Dataset/{sub}/*" 
    path_list = glob.glob(direc) 
    path_list = sorted( [pp.replace('\\', '/') for pp in path_list if '.' not in pp] ) 
#     print(path_list)
    all_paths += path_list
    
all_paths
```

```python
all_file_paths = [] 
for sub in all_paths:
    if '/subgroupII/' in sub:
        sub += '/1'
    direc = f"{sub}/*" 
    path_list = glob.glob(direc) 
    path_list = sorted( [pp.replace('\\', '/') for pp in path_list if '.txt' in pp] ) 
#     print(path_list)
    all_file_paths += path_list
    
all_file_paths
```

```python
len(all_paths), len(all_file_paths) 
```

```python
final_file_paths = [] 

for sub in all_file_paths:
    if '2.txt' in sub:
        continue 
    final_file_paths.append(sub)
    
final_file_paths
```
